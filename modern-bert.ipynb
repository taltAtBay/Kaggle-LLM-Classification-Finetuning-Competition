```python
#!/usr/bin/env python
import os
import torch
import torch.nn.functional as F
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dataclasses import dataclass
from tqdm import tqdm
from typing import List, Dict, Optional, Union, Callable


from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, classification_report

from torch.utils.data import Dataset
from torch.optim.lr_scheduler import OneCycleLR

from transformers import (
    AutoModelForSequenceClassification,
    AutoTokenizer,
    Trainer,
    TrainingArguments,
    DataCollatorWithPadding,
    EarlyStoppingCallback,
    TrainerCallback
)

# Optional: disable tokenizer parallelism to avoid warnings
os.environ["TOKENIZERS_PARALLELISM"] = "false"
torch.backends.cuda.matmul.allow_tf32 = True
torch.backends.cudnn.allow_tf32 = True

# ---------------------------
# Model configuration
# ---------------------------
@dataclass
class ModelConfig:
    model_name: str = "answerdotai/ModernBERT-base"  # HF model identifier
    max_length: int = 2048  # Maximum sequence length for truncation
    batch_size: int = 56
    learning_rate: float = 2e-5
    max_learning_rate: float = 1e-4 
    num_epochs: int = 5
    output_dir: str = "outputs"
    gradient_accumulation_steps: int = 4
    logging_steps: int = 25
    save_steps: int = 100
    eval_steps: int = 100
    torch_dtype: str = "bfloat16"  # switch to BF16
    drop_long_sequences: bool = True  # Drop samples that exceed max_length
    warmup_ratio = 0.2
    weight_decay: float = 0.01
    max_grad_norm = 1.0
    # lr_scheduler_type = "cosine"



config = ModelConfig()

# ---------------------------
# Custom Trainer with Label Smoothing
# ---------------------------
class CustomTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
        outputs = model(**inputs)
        logits = outputs.logits
        labels = inputs.get("labels")
        loss = F.cross_entropy(logits, labels, label_smoothing=0.1)
        return (loss, outputs) if return_outputs else loss

# ---------------------------
# Evaluation and Visualization
# ---------------------------
def evaluate_and_visualize(trainer, dataset, dataset_name=""):
    eval_result = trainer.predict(dataset)
    logits, labels = eval_result.predictions, eval_result.label_ids
    # Compute cross-entropy loss (without label smoothing) as our key metric
    loss_tensor = F.cross_entropy(torch.tensor(logits), torch.tensor(labels), reduction="mean")
    eval_cross_entropy_loss = loss_tensor.item()
    
    pred_labels = np.argmax(logits, axis=1)
    cm = confusion_matrix(labels, pred_labels)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {dataset_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    print(f"\nClassification Report - {dataset_name}")
    print(classification_report(labels, pred_labels))
    
    eval_result.metrics["eval_cross_entropy_loss"] = eval_cross_entropy_loss
    return eval_result.metrics

# ---------------------------
# Dataset Processing
# ---------------------------
class DatasetProcessor:
    def __init__(self, tokenizer_name: str, max_length: int, drop_long_sequences: bool = False):
        self.tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
        self.max_length = max_length
        self.separator = "\n\n###\n\n"
        self.drop_long_sequences = drop_long_sequences

    def construct_prompt(self, row: pd.Series) -> str:
        # Add explicit task instruction at start
        return f"Task: Compare two responses and choose the better one.\n\nPrompt: {row['prompt']}{self.separator}Response A: {row['response_a']}{self.separator}Response B: {row['response_b']}{self.separator}Choose: A/B/Tie"

    def get_label(self, row: pd.Series) -> int:
        # Derive label from winner columns:
        # 0 if winner_model_a == 1, 1 if winner_model_b == 1, 2 if winner_tie == 1.
        if row.get("winner_model_a") == 1:
            return 0
        elif row.get("winner_model_b") == 1:
            return 1
        elif row.get("winner_tie") == 1:
            return 2
        else:
            return -1  # Mark invalid label

    def check_sequence_length(self, prompt: str) -> bool:
        tokens = self.tokenizer(prompt, truncation=False)['input_ids']
        return len(tokens) <= self.max_length

    def pre_tokenize_data(self, df: pd.DataFrame) -> List[Dict[str, torch.Tensor]]:
        tokenized_data = []
        progress_bar = tqdm(total=len(df), desc="Tokenizing Dataset", unit="sample", dynamic_ncols=True)
        for _, row in df.iterrows():
            prompt = self.construct_prompt(row)
            if self.drop_long_sequences and not self.check_sequence_length(prompt):
                progress_bar.update(1)
                continue
            label = self.get_label(row)
            if label == -1:
                progress_bar.update(1)
                continue
            encoding = self.tokenizer(
                prompt,
                truncation=True,
                max_length=self.max_length,
                return_tensors="pt"
            )
            tokenized_data.append({
                "input_ids": encoding["input_ids"].squeeze(0),
                "attention_mask": encoding["attention_mask"].squeeze(0),
                "labels": torch.tensor(label, dtype=torch.long)
            })
            progress_bar.update(1)
        progress_bar.close()
        return tokenized_data

class PreTokenizedDataset(Dataset):
    def __init__(self, tokenized_data: List[Dict[str, torch.Tensor]]):
        self.tokenized_data = tokenized_data

    def __len__(self):
        return len(self.tokenized_data)

    def __getitem__(self, idx):
        item = self.tokenized_data[idx]
        return {
            'input_ids': item['input_ids'],
            'attention_mask': item['attention_mask'],
            'labels': item['labels']
        }

# ---------------------------
# Model Loading
# ---------------------------
def load_model(model_name: str, torch_dtype: str):
    model = AutoModelForSequenceClassification.from_pretrained(
        model_name,
        num_labels=3,
        torch_dtype=torch.bfloat16 if torch_dtype=="bfloat16" else torch.float16
    )
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.to(device)
    model.gradient_checkpointing_enable()
    return model

# ---------------------------
# Compute Metrics
# ---------------------------
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    acc = accuracy_score(labels, predictions)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, predictions, average='weighted')
    loss = F.cross_entropy(torch.tensor(logits), torch.tensor(labels), reduction="mean").item()
    return {
        'accuracy': acc,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        "eval_cross_entropy_loss": loss
    }
# -----------------------------------
# Callbacks
# ------------------------------------

class GradientDiagnosticCallback(TrainerCallback):
    def on_step_end(self, args, state, control, model=None, **kwargs):
        if state.global_step % 25 == 0:  # Log every 10 steps
            total_norm = 0.0
            param_stats = {}
            
            for name, param in model.named_parameters():
                if param.requires_grad and param.grad is not None:
                    param_norm = param.grad.data.norm(2).item()
                    total_norm += param_norm ** 2
                    
                    # Categorize parameters
                    if 'lora' in name:
                        param_type = 'LoRA'
                    elif 'score' in name:
                        param_type = 'Score'
                    else:
                        param_type = 'Other'
                    
                    if param_type not in param_stats:
                        param_stats[param_type] = {
                            'total_norm': 0.0,
                            'count': 0,
                            'max_norm': 0.0
                        }
                    
                    param_stats[param_type]['total_norm'] += param_norm
                    param_stats[param_type]['count'] += 1
                    param_stats[param_type]['max_norm'] = max(
                        param_stats[param_type]['max_norm'], 
                        param_norm
                    )
            
            total_norm = (total_norm ** 0.5)
            
            print(f"\nStep {state.global_step}:")
            print(f"Total Gradient Norm: {total_norm:.4f}")
            
            for param_type, stats in param_stats.items():
                if stats['count'] > 0:
                    print(f"{param_type} Parameters:")
                    print(f"  Count: {stats['count']}")
                    print(f"  Avg Norm: {stats['total_norm']/stats['count']:.4f}")
                    print(f"  Max Norm: {stats['max_norm']:.4f}")

class SimpleLoggingCallback(TrainerCallback):
    def __init__(self, log_file=None):
        self.log_file = log_file
        self.log_entries = []

    def _log(self, message):
        print(message)
        if self.log_file:
            self.log_entries.append(message)

    def on_train_begin(self, args, state, control, **kwargs):
        self._log("\n--- Training Started ---")
        self._log(f"Total Training Steps: {state.max_steps}")
        self._log(f"Batch Size: {args.per_device_train_batch_size}")
        self._log(f"Gradient Accumulation Steps: {args.gradient_accumulation_steps}")
        self._log(f"Learning Rate: {args.learning_rate}")

    def on_log(self, args, state, control, logs=None, **kwargs):
        if logs:
            log_str = " | ".join([f"{k}: {v:.4f}" for k, v in logs.items() if isinstance(v, (int, float))])
            self._log(f"Step {state.global_step}: {log_str}")

    def on_train_end(self, args, state, control, **kwargs):
        self._log("\n--- Training Completed ---")
        if self.log_file:
            with open(self.log_file, 'w') as f:
                f.write("\n".join(self.log_entries))


early_stopping_callback = EarlyStoppingCallback(
    early_stopping_patience=3,  # Number of evaluations to wait for improvement
    early_stopping_threshold=0.005  # Minimum change to qualify as an improvement
)

# ---------------------------
# Main Function
# ---------------------------
def main():
    # Load raw data from CSV (sample columns: id, model_a, model_b, prompt, response_a, response_b, winner_model_a, winner_model_b, winner_tie)
    df = pd.read_csv("data/train.csv")
    
    # Create a new "label" column from the winner columns:
    df["label"] = df.apply(lambda row: 0 if row["winner_model_a"] == 1 
                           else (1 if row["winner_model_b"] == 1 
                           else (2 if row["winner_tie"] == 1 else -1)), axis=1)
    # Remove rows with invalid label (-1)
    df = df[df["label"] != -1]
    
    # Split the data: 80% training; remaining 20% split equally into evaluation and test sets.
    train_df, temp_df = train_test_split(df, test_size=0.20, random_state=42, stratify=df["label"])
    eval_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42, stratify=temp_df["label"])
    
    print(f"Total samples: {len(df)}")
    print(f"Training samples (80%): {len(train_df)}")
    print(f"Evaluation samples (10%): {len(eval_df)}")
    print(f"Test samples (10%): {len(test_df)}")
    
    # Initialize the dataset processor with drop_long_sequences enabled.
    processor = DatasetProcessor(
        tokenizer_name=config.model_name,
        max_length=config.max_length,
        drop_long_sequences=config.drop_long_sequences
    )
    train_tokenized = processor.pre_tokenize_data(train_df)
    eval_tokenized = processor.pre_tokenize_data(eval_df)
    test_tokenized = processor.pre_tokenize_data(test_df)
    
    train_dataset = PreTokenizedDataset(train_tokenized)
    eval_dataset = PreTokenizedDataset(eval_tokenized)
    test_dataset = PreTokenizedDataset(test_tokenized)
    
    # Initialize the tokenizer for the collator.
    tokenizer = AutoTokenizer.from_pretrained(config.model_name)
    data_collator = DataCollatorWithPadding(tokenizer=tokenizer)  # Dynamic padding
    
    # Load the model.
    model = load_model(config.model_name, config.torch_dtype)
    
    
    # Setup TrainingArguments.
    training_args = TrainingArguments(
        output_dir=config.output_dir,
        per_device_train_batch_size=config.batch_size,
        per_device_eval_batch_size=config.batch_size,
        learning_rate=config.learning_rate,
        num_train_epochs=config.num_epochs,
        # lr_scheduler_type=config.lr_scheduler_type,
        # lr_scheduler_kwargs={"min_lr": 2e-6},
        warmup_ratio=config.warmup_ratio,
        eval_strategy="steps",   
        logging_strategy="steps",         
        save_strategy="steps", 
        load_best_model_at_end=True,
        fp16=False,      # disable fp16
        bf16=True,       # enable BF16 mixed precision
        gradient_accumulation_steps=config.gradient_accumulation_steps,
        logging_steps=config.logging_steps,
        eval_steps=config.eval_steps,
        save_steps=config.save_steps,
        max_grad_norm=config.max_grad_norm,
        weight_decay=config.weight_decay,
        metric_for_best_model="eval_cross_entropy_loss",
        greater_is_better=False,
        logging_dir="./logs",
        report_to="none",
        group_by_length=True,
        # Added gradient checkpointing
        gradient_checkpointing=True,
        save_total_limit=2,
    )
    
    # Initialize our custom trainer.
    trainer = CustomTrainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=eval_dataset,
        compute_metrics=compute_metrics,
        data_collator=data_collator,
        callbacks=[early_stopping_callback, GradientDiagnosticCallback, SimpleLoggingCallback(log_file=f"{config.output_dir}/training_log.txt")]
    )

    # Force creation of the optimizer
    if trainer.optimizer is None:
        trainer.create_optimizer()
    
    total_steps = (len(train_dataset) // (config.batch_size * config.gradient_accumulation_steps) + 1) * config.num_epochs
    scheduler = OneCycleLR(
        trainer.optimizer,
        max_lr=config.max_learning_rate,
        total_steps=total_steps,
        pct_start=0.3,  # Spend 30% of iterations in increasing LR
        div_factor=25,  # Initial learning rate is max_lr/25
        final_div_factor=1e4,  # Final learning rate is initial_lr/10000
        anneal_strategy='cos',
        cycle_momentum=True,
        base_momentum=0.85,
        max_momentum=0.95
    )

    trainer.lr_scheduler = scheduler
    
    print("Starting training...")
    trainer.train()
    
    print("Evaluating on evaluation set...")
    eval_metrics = evaluate_and_visualize(trainer, eval_dataset, dataset_name="Evaluation Set")
    print("Evaluation Metrics:", eval_metrics)
    
    print("Evaluating on test set...")
    test_metrics = evaluate_and_visualize(trainer, test_dataset, dataset_name="Test Set")
    print("Test Metrics:", test_metrics)
    
    # Save the best model.
    trainer.save_model("./best_model")

if __name__ == "__main__":
    main()
```
